```python
def extended_euclicid(a, b):
    """ Extended Euclicid Algorithm, returns gcd, and x, y in d = gcd(a, b) = ax + by
        complexity: O(lg(b))"""
    if b == 0:
        return [a, 1, 0]
    else:
        [dn, xn, yn] = extended_euclicid(b, a % b)
        [d, x, y] = [dn, yn, xn - (a // b) * yn]
        return [d, x, y]

def modular_linear_eq_solver(a, b, n):
    result = []
    [d, xp, yp] = extended_euclicid(a, n)
    if b % d == 0:
        x0 = (xp * (b // d)) % n
        for i in range(0, d):
            result.append((x0 + i*(n // d)) % n)
    else:
        result = ["no solutions"]
    return result
import time
tm = []
for x in range(1, 9999999, 100000):
	t0 = time.time()
	for _ in range(1000):
		modular_linear_eq_solver(512553777, 1, x)
	t1 = time.time()
	tm.append(t1 - t0)
```
